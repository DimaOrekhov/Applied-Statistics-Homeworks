# Корреляционный анализ файла Adstudy.sta

Импортируем необходимые модули:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=[10,7]
import seaborn as sns
sns.set()
import scipy.stats as st
```

Считываем данные:

```python
data=pd.read_excel('adstudy.xlsx', index_col=0, skiprows=1)
data.head()
```

Выделяем колонки MEASURE01-MEASURE015. И строим для них корреляционную матрицу:

```python
measures=['MEASURE0%s'%(x) if x<10 else 'MEASURE%s'%(x) for x in range(1, 16)]
data[measures].corr()
```

Построим график heatmap для корреляций, интенсивность цвета соответсвует значениям корреляций.

```python
plt.figure(figsize=[12, 8])
sns.heatmap(data[measures].corr())
```

Преобразуем график так, чтобы незначимые корреляции (уровень значимости 0.05) отображались белым, как и корреляции переменных с самими собой:

```python
corr_mat=np.zeros([15, 15])
for i in range(len(measures)):
    for j in range(len(measures)):
        r, p_val=st.pearsonr(data[measures].iloc[:,i], data[measures].iloc[:,j])
        if p_val>0.05:
            corr_mat[i, j]=float(1)
        else:
            corr_mat[i, j]=float(r)

plt.figure(figsize=[12, 8])
_=sns.heatmap(corr_mat, yticklabels=measures, xticklabels=measures)
```

Видно, что значимы корреляции лишь для следующих пар:

- MEASURE03 - MEASURE15
- MEASURE05 - MEASURE09
- MEASURE06 - MEASURE08
- MEASURE07 - MEASURE13
- MEASURE10 - MEASURE15

Для них можно отвергнуть нулевую гипотезу о равенстве коэффициента корреляции нулю, приведем значения коэффициентов для этих пар в ниже:

```python
m_1=['MEASURE0%s'%(x) if x<10 else 'MEASURE%s'%(x) for x in [3,5,6,7,10]]
m_2=['MEASURE0%s'%(x) if x<10 else 'MEASURE%s'%(x) for x in [15,9,8,13,15]]
for x_1, x_2 in zip(m_1, m_2):
    r,p=st.pearsonr(data[x_1], data[x_2])
    n_s='    '
    p_s='     '
    if r<0 and np.round(r, decimals=3)!=-0.33:
        print(x_1+'-'+x_2+': r='+str(np.round(r, decimals=3))+n_s+'p_val='+str(np.round(p, decimals=3)))
    else:
        print(x_1+'-'+x_2+': r='+str(np.round(r, decimals=3))+p_s+'p_val='+str(np.round(p, decimals=3)))
```
